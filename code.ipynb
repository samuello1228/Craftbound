```python
from math import floor, ceil


def XP_Per_Rotation(recipe, level):
    if level < recipe["Level"]:
        print("Error: the level is lower than the required level.")

    XP_base = recipe["Base XP"]
    level_diff = level - recipe["Level"]
    XP = XP_base * (1 - level_diff * 0.1) * 4
    XP = floor(XP)
    XP = max(XP, 20)

    return XP
```

```python
player_Progress = 389
player_Quality = 190
player_CP = 143


def find_ingredient(production_chain, Item_Name, Rarity):
    # Find node for the ingredient
    for i in range(len(General_Goods)):
        if General_Goods[i]["Item Name"] == Item_Name:
            return (-1, i)

    for i in range(len(production_chain)):
        for j in range(len(production_chain[i]["products"])):
            if production_chain[i]["products"][j]["item_key"] == (Item_Name, Rarity):
                return (i, j)

    return None


def build_production_chain():
    production_chain = []

    for skill in Skills:
        if skill["Skill Type"] == "Gathering":
            for location in skill["Data"]:
                node = {}
                node["n_rotation"] = 0
                node["Item Name"] = location["Drops"][0]["Item Name"]
                node["ingredients"] = []

                node["products"] = []
                for drop in location["Drops"]:
                    product = {}
                    product["item_key"] = (drop["Item Name"], drop["Rarity"])

                    if drop["Rarity"] == "Common":
                        Quantity_per_Rotation = 4
                    else:
                        Quantity_per_Rotation = 1

                    product["Rotation_per_Quantity"] = 1 / (Quantity_per_Rotation * drop["Chance"])

                    product["Quantity_required"] = 0

                    node["products"].append(product)

                node["time"] = location["Wait Time"]
                node["time"] += 6

                production_chain.append(node)

        elif skill["Skill Type"] == "Crafting":
            for recipe in skill["Data"]:
                node = {}
                node["n_rotation"] = 0
                node["Item Name"] = recipe["Item Name"]

                node["ingredients"] = []
                for ingredient in recipe["Ingredients"]:
                    index = find_ingredient(production_chain, ingredient["Item Name"], ingredient["Rarity"])
                    if index == None:
                        print("Error: Ingredient cannot be found in the previous nodes.")

                    item_key = (ingredient["Item Name"], ingredient["Rarity"])
                    node["ingredients"].append({"item_key": item_key,
                                                "Amount": ingredient["Amount"],
                                                "node_index": index})

                node["products"] = []
                sum_of_chance = 0
                for rarity in recipe["Rarity"]:
                    product = {}
                    product["item_key"] = (recipe["Item Name"], rarity)

                    if rarity == "Common":
                        product["Chance"] = 1
                    elif rarity == "Uncommon":
                        product["Chance"] = 0.1
                    elif rarity == "Rare":
                        product["Chance"] = 0.001
                    elif rarity == "Epic":
                        product["Chance"] = 0.0001
                    elif rarity == "Legendary":
                        product["Chance"] = 0.00001

                    sum_of_chance += product["Chance"]

                    node["products"].append(product)

                # normalize chance
                for product in node["products"]:
                    product["Chance"] /= sum_of_chance
                    product["Rotation_per_Quantity"] = 1 / product["Chance"]
                    del product["Chance"]

                    product["Quantity_required"] = 0

                node["time"] = 1
                node["time"] += ceil(recipe["Crafting Progress"]/player_Progress)
                node["time"] += ceil(recipe["Crafting Quality"]/player_Quality)
                production_chain.append(node)

    return production_chain
```

```python
def measure_time(settings, production_chain):
    # fill node["n_rotation"]
    for setting in settings:
        for product in setting:
            for node in production_chain:
                if node["Item Name"] == product["Item Name"]:
                    # print(node["Item Name"])
                    node["n_rotation"] = product["Quantity"]
                    break

    time_used = 0
    for node in reversed(production_chain):
        n_rotation_max = node["n_rotation"]
        node["n_rotation"] = 0

        for product in node["products"]:
            Rotation_required = product["Quantity_required"] * product["Rotation_per_Quantity"]
            n_rotation_max = max(Rotation_required, n_rotation_max)
            product["Quantity_required"] = 0

        if n_rotation_max == 0:
            continue

        time_used += n_rotation_max * node["time"]

        for ingredient in node["ingredients"]:
            (i, j) = ingredient["node_index"]
            # print((i, j))
            if i == -1:
                continue
            production_chain[i]["products"][j]["Quantity_required"] += n_rotation_max * ingredient["Amount"]

    return time_used
```

```python
def check_minimum_level(setting, partition_per_level):
    nPartition_accumulated = 0
    for i in range(len(setting)):
        if nPartition_accumulated < (setting[i]["Recipe"]["Level"] - 1) * partition_per_level:
            return False

        nPartition_accumulated += setting[i]["nPartition"]
    return True
```

```python
def try_new_settings(settings, production_chain, setting, time_used, partition_per_level, delta):
    isFaster = False
    i_best = None
    j_best = None
    for i in range(len(setting)):
        if setting[i]["nPartition"] - delta < 0:
            continue

        # decrement
        setting[i]["nPartition"] -= delta

        for j in range(len(setting)):
            if i == j:
                continue

            # increment
            setting[j]["nPartition"] += delta

            if not check_minimum_level(setting, partition_per_level):
                # undo increment
                setting[j]["nPartition"] -= delta
                continue

            fill_Quantity(setting, partition_per_level)
            time_used_new = measure_time(settings, production_chain)

            # print(setting[i]["Item Name"] + ": ", setting[i]["nPartition"],
            #       ", ", setting[j]["Item Name"] + ": ", setting[j]["nPartition"], sep="")
            # print("time_used_new: ", time_used_new, sep="")

            if time_used - time_used_new > 1:
                # print(setting[i]["Item Name"] + ": ", setting[i]["nPartition"] + delta, " => ", setting[i]["nPartition"],
                #       ", ", setting[j]["Item Name"] + ": ", setting[j]["nPartition"] - delta, " => ", setting[j]["nPartition"], sep="")
                # print("Faster: ", time_used, " => ", time_used_new, sep="")

                isFaster = True
                i_best = i
                j_best = j
                time_used = time_used_new

            # undo increment
            setting[j]["nPartition"] -= delta

        # undo decrement
        setting[i]["nPartition"] += delta

    if isFaster:
        setting[i_best]["nPartition"] -= delta
        setting[j_best]["nPartition"] += delta

    fill_Quantity(setting, partition_per_level)

    if isFaster:
        print("delta:", delta)
        print(setting[0]["Skill Name"])
        print(setting[i_best]["Item Name"] + ": ", setting[i_best]["nPartition"] + delta, " => ", setting[i_best]["nPartition"], ", ",
              setting[j_best]["Item Name"] + ": ", setting[j_best]["nPartition"] - delta, " => ", setting[j_best]["nPartition"], sep="")
        print("Best time:", time_used)
        print("Best setting:")
        for product in setting:
            print(product)
        print()

    return (isFaster, time_used)
```

```python
def coarse_search(settings, production_chain, max_level, partition_per_level):
    time_used = measure_time(settings, production_chain)
    for delta in reversed(range(1, max_level+1)):
        while True:
            isChanged = False

            for setting in settings:
                (isFaster, time_used) = try_new_settings(settings,
                                                         production_chain, setting, time_used, partition_per_level, delta)
                if isFaster:
                    isChanged = True

            if not isChanged:
                break

    return settings
```

```python
def fine_search(settings, production_chain, partition_per_level):
    time_used = measure_time(settings, production_chain)
    while True:
        isChanged = False

        for setting in settings:
            (isFaster, time_used) = try_new_settings(settings,
                                                     production_chain, setting, time_used, partition_per_level, 1)
            if isFaster:
                isChanged = True

        if not isChanged:
            break

    return settings
```
